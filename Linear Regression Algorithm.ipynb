```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
```

```python
df=pd.read_csv("Scores.csv")
```

```python
df
```

```python
df.info()
```

```python
#Separation of X and Y x=Feature and y=Target
x=df[['Hours']]
y=df[['Scores']]
```

```python
x
```

```python
y
```

```python
#Splitting into train and test
from sklearn.model_selection import train_test_split #here train_test_split is class
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.2,random_state=0) #random_state=0 means to fix the splitting done-permanent changes(0.2 means the data sent for testing is 20%-xtest,ytest and the data sent for training is 80%-xtrain,ytrain.
#here the ratio of train:test data is 80:20.it can also be 70:30,60:40 but always remember the data sent for training is always larger then the test data.
```

```python
xtrain
```

```python
xtrain
```

```python
#Build model
#Step 1:Import model
from sklearn.linear_model import LinearRegression   #sklearn-library linear-model=module
#Step 2:Create instance of model
lr=LinearRegression()
#Step 3:train a model
lr.fit(xtrain,ytrain)
#Step 4: Predict values
ypred=lr.predict(xtest)
```

```python
ypred
```

```python
#Evaluate Model(Checking accuracy)
from sklearn.metrics import r2_score
r2_score(ytest,ypred)
```

```python
#To know best fit line slope value
lr.coef_
```

```python
#To know best fit line intercept value
lr.intercept_
```

```python
#To know user defined prediction score
lr.predict([[10.2]]) #give hours(xtest)
```
